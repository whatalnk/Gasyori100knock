```
import matplotlib.pyplot as plt
import numpy as np
```

左上を黒に

```
img2 = img.copy()
img2[:50, :50] = 0
cv2_imshow(img2)
```

260 - 256 で黄色になる

```
img2 = img.copy().astype(np.float32)
img2[60:100, 60:100, 0] = 260
cv2_imshow(img2.astype(np.uint8))
```

```
cv2.imwrite("sample.jpg", img2)
```

## 練習

画像の左半分上のR と B を入れ替え

```
img3 = img.copy()
H, W, C = img3.shape
print((H, W, C))
```

```
img3[:H//2, :W//2] = img3[:H//2, :W//2, (2, 1, 0)] # B G R
cv2_imshow(img3)
```
